# **Real Estate price Prediction**

## **Load Data**

```python
import pandas as pd
```

```python
df=pd.read_csv('/Users/rayquaza/Desktop/Real estate price prediction/dataset/Real estate.csv',index_col=0)
df
```

```python
df.count()
```

```python
df.columns
```

## **Data cleanning**

### Sepration of x and y

```python
y=df['Y house price of unit area']
y
```

```python
x=df.drop('Y house price of unit area',axis=1)
x
```

### Data Splitting

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2,random_state=100)
```

```python
x_train
```

```python
x_test
```

## **Model building**

### Linear regression

```python
from sklearn.linear_model import LinearRegression 
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

### Applying the model

```python
y_lr_pred_train=lr.predict(x_train)
y_lr_pred_test=lr.predict(x_test)
```

### Evaluate the model

```python
from sklearn.metrics import mean_squared_error, r2_score 
```

```python
y_lr_pred_test_mse=mean_squared_error(y_test,y_lr_pred_test)
y_lr_pred_train_mse=mean_squared_error(y_train,y_lr_pred_train)


y_lr_pred_test_r2=r2_score(y_test,y_lr_pred_test)
y_lr_pred_train_r2=r2_score(y_train,y_lr_pred_train)
```

```python
results_lr=pd.DataFrame(['Ljnear Regression',y_lr_pred_test_mse,y_lr_pred_train_mse,y_lr_pred_test_r2,y_lr_pred_train_r2]).transpose()
results_lr.columns=['Model','Test MSE','Train MSE','Test R2','Train R2']
results_lr
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
```

### Applying the model 

```python
y_rf_pred_train=lr.predict(x_train)
y_rf_pred_test=lr.predict(x_test)
```

### Evaluating the model

```python
from sklearn.metrics import mean_squared_error ,r2_score
```

```python
y_rf_pred_test_mse=mean_squared_error(y_test,y_rf_pred_test)
y_rf_pred_train_mse=mean_squared_error(y_train,y_rf_pred_train)


y_rf_pred_test_r2=r2_score(y_test,y_rf_pred_test)
y_rf_pred_train_r2=r2_score(y_train,y_rf_pred_train)
```

```python
results_rf=pd.DataFrame(['Ljnear Regression',y_rf_pred_test_mse,y_rf_pred_train_mse,y_rf_pred_test_r2,y_rf_pred_train_r2]).transpose()
results_rf.columns=['Model','Test MSE','Train MSE','Test R2','Train R2']
results_rf
```

# **Data Visualization of prediction**

## concatination

```python
results=pd.concat([results_lr,results_rf])
results
```

### Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Linear Regression

```python
plt.scatter(x=y_test,y=y_lr_pred_test,alpha=0.3)
plt.xlabel('Experimental Values')
plt.ylabel('Predicted Values')
z=np.polyfit(y_train,y_lr_pred_train,1)
p=np.poly1d(z)
plt.plot(y_train,p(y_train),'Red')
```

#### Random Forest

```python
plt.scatter(x=y_test,y=y_rf_pred_test,alpha=0.3)
plt.xlabel('Experimental Values')
plt.ylabel('Predicted Values')
z=np.polyfit(y_train,y_rf_pred_train,1)
p=np.poly1d(z)
plt.plot(y_train,p(y_train),'Red')
```
